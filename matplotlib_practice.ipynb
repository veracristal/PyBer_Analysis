```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)


# Create a title
plt.title("PyBer Fare by Month")

#Add a grid.
plt.grid()

# Add the legend
plt.legend()

```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)

# Create the plot and add a label for the legend.
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)


# Create a title
plt.title("PyBer Fare by Month")

#Add a grid.
plt.grid()

# Add the legend
plt.legend()

plt.show()
```

```python
# Create a Bubble chart using the Object-Oriented Approach
fig, ax = plt.subplots()
# Bubble chart making markers 5 times bigger
y_axis_larger =[]
for data in y_axis:
    y_axis_larger.append(data*5)

ax.scatter(y_axis, x_axis, s=y_axis_larger, color='skyblue', alpha=.3, edgecolors='black', linewidths=2, label="Boston")

# invert labels so they read top to bottom instead of bottom to top
ax.invert_yaxis()

# Set labels and title
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')



# Add the legend
plt.legend()

```

```python
# Create a Pie Chart using Object-Oriented interface
fig, ax = plt.subplots()


# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)


ax.pie(y_axis, labels=x_axis, autopct='%.1f%%', colors=colors, explode=explode_values, shadow=True, startangle=270, radius=2)

# invert labels so they read top to bottom instead of bottom to top
ax.invert_yaxis()

#ax.subplots(figsize=(8, 8))







plt.show()
```
